```python
students=["Ravi","Anita","Ravi","Suresh","Anita"]
students.append("Kiran")
print("After adding new student:",students)
backup=students.copy()
print("Backup copy:",backup)
name=input("Enter student name to count:")
print("Count:",students.count(name))
transfer_students=["Meena","Rahul"]
students.extend(transfer_students)
print("After merging transfer students:",students)
search=input("Enter student name to find position:")
if search in students:
 print("Position:",students.index(search))
else:
 print("Student not found")
priority=input("Enter priority student name:")
pos=int(input("Enter position:"))
students.insert(pos,priority)
print("After inserting priority student:",students)
removed_last=students.pop()
print("Removed last enrolled student:",removed_last)
withdraw=input("Enter withdrawn student name:")
if withdraw in students:
 students.remove(withdraw)
 print("After removing withdrawn student:",students)
else:
 print("Student not found")
print("Reverse order:",students[::-1])
students.sort()
print("Sorted list:",students)
students.clear()
print("All records cleared:",students)
```
