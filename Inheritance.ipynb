```python
class Person:
  def __init__(self,name):
    self.name=name
  def display(self):
    print(self.name)
class Employee(Person):
  def __init__(self,name):
    Person.__init__(self,name)
class Manager(Employee):
   def __init__(self,name,age):
    Employee.__init__(self,name)
    self.age=age
p=Employee("ram")
p.display()
q=Manager("sam",30)
print(q.age)
class Student(Person):
  def __init__(self,name,degree):
    Person.__init__(self,name)
    self.degree=degree
  def display(self):
    print(self.name,self.degree)
class SportsStudent(Student):
  def __init__(self,name,degree,sport):
    Student.__init__(self,name,degree)
    self.sport=sport
  def display(self):
    print(self.name,self.degree,self.sport)
class AcademicStudent(Student):
  def __init__(self,name,degree,semester):
    Student.__init__(self,name,degree)
    self.semester=semester
  def display(self):
    print(self.name,self.degree,self.semester)
r=Student("ram","btech")
r.display()
s=SportsStudent("ram","btech","cricket")
s.display()
t=AcademicStudent("ram","btech",3)
t.display()
class CollegeStudent(SportsStudent):
  def __init__(self,name,degree,sport,college):
    SportsStudent.__init__(self,name,degree,sport)
    self.college=college
  def display(self):
    print(self.name,self.degree,self.sport,self.college)
u=CollegeStudent("ram","btech","cricket","srm")
u.display()


```
