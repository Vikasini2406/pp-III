```python
books={
1:{"title":"Physics","author":"Halliday","available_copies":5},
2:{"title":"Maths","author":"Thomas","available_copies":3},
3:{"title":"Chemistry","author":"Brown","available_copies":4}
}
backup=books.copy()
print("Backup:",backup)
categories=dict.fromkeys(["Science","Arts","Engineering"],0)
print("Categories:",categories)
bid=int(input("Enter book id:"))
print("Available copies:",books.get(bid,{}).get("available_copies","Not found"))
print("All book records:")
for k,v in books.items():
 print(k,v)
print("Book IDs:",list(books.keys()))
copies=[]
for v in books.values():
 copies.append(v["available_copies"])
print("Available copies list:",copies)
removed=books.pop(2)
print("Removed book:",removed)
last=books.popitem()
print("Last removed:",last)
categories.setdefault("Medical",0)
print("Updated categories:",categories)
new_books={
4:{"title":"Biology","author":"Watson","available_copies":6}
}
books.update(new_books)
print("After merging:",books)
books.clear()
print("All records cleared:",books)
```

```python
registered_students={101,102,103,104}
workshop_A={101,103,105}
workshop_B={102,103,106}
volunteers={101,104}
registered_students.add(105)
backup=registered_students.copy()
print("Backup:",backup)
print("Registered but not in workshop A:",registered_students.difference(workshop_A))
cancelled={104}
registered_students.difference_update(cancelled)
print("After removing cancelled:",registered_students)
registered_students.discard(200)
print("After discard:",registered_students)
print("Workshops disjoint?",workshop_A.isdisjoint(workshop_B))
print("Volunteers subset of registered?",volunteers.issubset(registered_students))
print("Students in both workshops:",workshop_A.intersection(workshop_B))
workshop_A.intersection_update(workshop_B)
print("Workshop A after intersection_update:",workshop_A)
removed_volunteer=volunteers.pop()
print("Removed volunteer:",removed_volunteer)
immutable_group=frozenset({101,102})
print("Immutable group:",immutable_group)
print("Students in exactly one workshop:",workshop_A.symmetric_difference(workshop_B))
all_participants=workshop_A.union(workshop_B)
print("All participants merged:",all_participants)
workshop_A.clear()
workshop_B.clear()
volunteers.clear()
print("All temporary data cleared:",workshop_A,workshop_B,volunteers)
```

```python
def count_vowels(s):
 s=s.lower()
 count=0
 for char in s:
  if char in "aeiou":
   count+=1
 return count
text=input("Enter a string:")
print("Number of vowels:",count_vowels(text))
```

```python
students={"Anu":[78,85,90],"Bala":[60,72,68],"Charan":[88,91,84],"Divya":[45,50,40]}
def calculate_average(marks):
 return sum(marks)/len(marks)
is_pass=lambda avg:"Pass" if avg>=50 else "Fail"
student_report={}
for name,marks in students.items():
 avg=round(calculate_average(marks),2)
 result=is_pass(avg)
 student_report[name]={"Average":avg,"Result":result}
print("Passed students:")
for name,details in student_report.items():
 if details["Result"]=="Pass":
  print(name,details)
topper=max(student_report.items(),key=lambda x:x[1]["Average"])
print("Topper:",topper[0])
```

```python
def count_digits(n):
 if n==0:
  return 0
 return 1+count_digits(n//10)
def sum_digits(n):
 if n==0:
  return 0
 return n%10+sum_digits(n//10)
def max_digit(n):
 if n==0:
  return 0
 return max(n%10,max_digit(n//10))
def reverse_number(n,rev=0):
 if n==0:
  return rev
 return reverse_number(n//10,rev*10+n%10)
num=int(input("Enter a positive integer:"))
print("Number of digits:",count_digits(num))
print("Sum of digits:",sum_digits(num))
print("Maximum digit:",max_digit(num))
print("Reversed number:",reverse_number(num))
```
