```python
class Circle:
  def __init__(self, radius):
    self.radius = radius
  def getArea(self):
    return 3.14 * self.radius * self.radius
  def getPerimeter(self):
    return 2 * 3.14 * self.radius
a1 = Circle(5)
print(a1.getArea())
print(a1.getPerimeter())
```

```python
class Vehicle:
  def __init__(self,name,colour,cost):
    self.name=name
    self.colour=colour
    self.cost=cost
  def display(self):
    print(self.name,self.colour,self.cost)
car1=Vehicle("Fer","red",60000)
car1.display()
car2=Vehicle("jump","blue",10000)
car2.display()
```

```python
class Temprature:
  def __init__(self,temp):
    self.temp=temp
  def convertFahrenheit(self):
    return (self.temp*9/5)+32
  def covertCelcius(self):
    return (self.temp-32)*5/9
temp1=Temprature(37)
print(temp1.convertFahrenheit())
print(temp1.covertCelcius())
```

```python
class Student:
  def __init__(self,name,rollno):
    self.name=name
    self.rollno=rollno
  def display(self):
    print(self.name,self.rollno)
  def setAge(self):
    self.age=20
    print(self.age)
  def setMarks(self):
    self.marks=100
    print(self.marks)
s1=Student("ram",1)
s1.display()
s1.setAge()
s1.setMarks()
```

```python
class Time:
    def __init__(self, hours, minutes):
        self.hours = hours
        self.minutes = minutes
    def addTime(t1, t2):
        total_hours = t1.hours + t2.hours
        total_minutes = t1.minutes + t2.minutes
        total_hours += total_minutes // 60
        total_minutes = total_minutes % 60
        return Time(total_hours, total_minutes)
    def displayTime(self):
        print(f"{self.hours} hr {self.minutes} min")
    def displayMinute(self):
        total = self.hours * 60 + self.minutes
        print("Total Minutes:", total)
t1 = Time(2, 50)
t2 = Time(1, 20)
result = Time.addTime(t1, t2)
print("Added Time:")
result.displayTime()
t3 = Time(1, 2)
t3.displayMinute()
```

```python
class Triangle:
  def __init__(self,angle1,angle2,angle3):
    self.angle1=angle1
    self.angle2=angle2
    self.angle3=angle3
    self.number_of_sides=3
  def check_angles(self):
    if self.angle1+self.angle2+self.angle3==180:
      return True
    else:
      return False
my_triangle=Triangle(30,60,90)
print(my_triangle.check_angles())
print(my_triangle.number_of_sides)
```

```python
class Songs:
  def __init__(self,lyrics):
    self.lyrics=lyrics
  def sing_me_a_song(self):
    for line in self.lyrics:
      print(line)
happy_bday=Songs(["May god bless you,","Have a sunshine","happy bday to you!"])
happy_bday.sing_me_a_song()
```
